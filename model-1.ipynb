```python
import numpy as np 
import pandas as pd # data processing, 
import matplotlib.pyplot as plt # data visualization
import seaborn as sns # statistical data visualization
```

```python
b15_cse=pd.read_csv("/home/dulangi/Desktop/miniData.csv")
b15_cse
```

```python
X_sem7 = b15_cse.drop(['g8','class'], axis=1)
X_sem7
```

```python
y=b15_cse['class']
y
```

```python
# import DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
```

```python
# instantiate the DecisionTreeClassifier model with criterion gini index

clf_gini_max_dep5 = DecisionTreeClassifier(criterion='gini', max_depth=5, random_state=0)
```

```python
# instantiate the DecisionTreeClassifier model with criterion entropy

clf_en_max_dep5  = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=0)
```

```python
asc_list = ['FirstClass',
            'Pass', 
            'SecondLower',  
            'SecondUpper']
```

```python
asc_list
```

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train_sem7, X_test_sem7, y_train_sem7, y_test_sem7 = train_test_split(X_sem7, y, test_size = 0.33, random_state = 42)
```

```python
feature_names_sem7=list(X_sem7.columns)
```

```python
from sklearn import tree
```

```python
# sem 5
plt.figure(figsize=(20,9))
tree.plot_tree(clf_gini_max_dep5.fit(X_train_sem7, y_train_sem7), filled=True,rounded=True,feature_names=feature_names_sem7,class_names=asc_list); 
```

```python
# sem 5
plt.figure(figsize=(20,9))
tree.plot_tree(clf_gini_max_dep5.fit(X_train_sem7, y_train_sem7), filled=True,rounded=True,feature_names=feature_names_sem7,class_names=asc_list); 
```

```python
# sem 5
plt.figure(figsize=(20,9))
tree.plot_tree(clf_en_max_dep5.fit(X_train_sem7, y_train_sem7), filled=True,rounded=True,feature_names=feature_names_sem7,class_names=asc_list); 
```
